```rust
:dep polars = { version = "0.25.1" , features = ["lazy", "csv-file", "strings", "temporal", "dtype-duration", "dtype-categorical", "concat_str" ]}
```

```rust
use polars::prelude::*;
```

```rust
let df:LazyFrame = CsvReader::from_path("/Users/squeeko/AI-for-CyberSecurity/JA_Port_to_Rust_Notebooks/data/ddos_dataset.csv")?
    .has_header(true)
    .finish()?
    .lazy();
```

```rust
//Select 5 rows from the dataset using .fetch(#)?, this like head() in pandas
df.clone()
    .select([col("*")
    ])
 .fetch(5)? 

    
```

```rust
//Selects the whole dataset using .collect()?
df.clone()
    .select([col("*")
    ])
 .collect()? 

    
```

```rust
//Select certain whole columns usign collect()? 
df.clone()
        .select([
        col("Flow ID"),
        col("Init Bwd Win Byts"),
])
.collect()?
```

```rust
//Filter columns based on a condition using collect() to show result
df.clone()
.filter(col("Init Bwd Win Byts")
.gt(lit(225)))
.collect()
```

```rust
//Filter columns based on a condition and sorted using SortOptions and collect() to show result
df.clone()
.filter(col("Init Bwd Win Byts")
.gt(lit(225))) //greater than and there are other like .lt etc. also you have to use lit() for literal num/string
.sort("Init Bwd Win Byts", SortOptions{descending: true, nulls_last:false})
//.sort("Init Bwd Win Byts", SortOptions::default()) - default sort
.collect()
```

```rust
// Aggregate using .groupby and .agg to find the min, max, median, mean and sum of the labels
df.clone()
.groupby(["Label"])
    .agg([
        col("Init Bwd Win Byts").min().alias("IBWB_MIN"),
        col("Init Bwd Win Byts").max().alias("IBWB_MAX"),
        col("Init Bwd Win Byts").median().alias("IBWB_MEDIAN"),
        col("Init Bwd Win Byts").mean().alias("IBWB_MEAN"),
        col("Init Bwd Win Byts").sum().alias("IBWB_SUM"),
]).collect()
```
